对于经营者来说，最需要了解的就是以下几点：
1.哪家店销售情况最好。
2.哪些商品畅销。
3.购物者的选择偏好，即购买某一商品的顾客大部分会同时购买什么商品，形成推荐。
4.怎么推荐一些折扣商品吸引更多顾客办理会员。

```python
import pandas as pd
data1 = pd.read_csv(r'Basket-2018-06-01-2018-07-05.csv')
data1.head()
```

```python
data1.shape
```

可以看出这个数据集有近60万条数据

```python
data1["SHOP_DATETIME"] = pd.to_datetime(data1["SHOP_DATETIME"])
```

我们首先了解一下这一个月内哪家店总销量最高。

```python
df1=data1.groupby('STORE_CODE').sum()
df1
```

```python
df1.sort_values(by=['BASK_SPEND'])
```

用图形进行更为直观的表示

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
y=['A001','A004','A006','A007','A008','A009','A010','A012','A013','A016','A017','A018','A019','A021','A022','D002','D003','D004','D005','D006','D007','D008','D009','D010']
height=df1['BASK_SPEND']
plt.grid(axis="y",which="major")
plt.xlabel('销售额(元)')
plt.ylabel('门店')
plt.title('2018年6月各门店销售情况分析图')
plt.barh(y,height,align='center',color='b',alpha=0.5)
plt.show()
```

可以看到，销售量最高的是A001,销量最低的是D003。对门店进行分析之后，我们对商品进行具体分析。

用户消费金额、消费次数的描述统计

```python
# 客户消费分析
grouped_Customer = data1[['BASK_CODE','SHOP_DATETIME', 
          'BASK_SPEND']].sort_values(['SHOP_DATETIME']).groupby('BASK_CODE')
grouped_Customer.head()
```

```python
# 用户消费金额、消费次数的描述统计(在此为平均每个用户平均消费金额，不是每笔平均消费金额)
grouped_Customer.count().describe()
```

```python
grouped_Customer['BASK_SPEND'].sum().describe()
```

```python
# 用户消费金额分布图
grouped_Customer.sum().BASK_SPEND.plot.hist(bins = 20)
```

可以看出用户消费金额大都集中在0-1000元以内,从上述直方图可以看出，用户消费金额，绝大部分是呈现集中趋势.

```python
#使用 plt 的 .boxplot() 方法绘制箱线图
plt.figure()
plt.subplot()
plt.boxplot(data1["BASK_SPEND"])
```

```python
# 用户累计消费占比（百分之多少的用户占百分之多少的销售额）
grouped_Customer.sum().sort_values('BASK_SPEND').apply(lambda x : x.cumsum()/ x.sum()).reset_index().BASK_SPEND.plot()
```

```python
# 用户第一次消费（首购）
grouped_Customer.min().SHOP_DATETIME.value_counts().plot()
```

```python
# 用户最后一次消费
grouped_Customer.max().SHOP_DATETIME.value_counts().plot()
```

新老用户消费比

```python
# 多少用户仅消费1次
user_life =grouped_Customer.SHOP_DATETIME.agg(['min','max'])
(user_life['min'] == user_life['max']).value_counts()
# 可视化（饼图）
plt.pie((user_life['min'] == user_life['max']).value_counts(normalize = True),
       autopct='%.2f%%',
       labels = (user_life['min'] == user_life['max']).value_counts().index)
```
