 <font color=black size=4>&emsp;&emsp;首先导入numpy包和pandas方便下一步对数据集操作分析，len函数可以查看交易明细数据的总体交易记录，如下显示2018年6月1日-2018年7月5日各门店超市共有交易记录1732761条。item表显示交易明细共有10个指标来记录交易情况。

### 1.交易数量问题

<font color=black size=4>&emsp;&emsp;如下表所示，超市交易明细中记录2022条的交易数目存在问题，问题在于交易数量为负值。按照实际交易情况分析，交易数目应为非负数，因此本报告将2022条明细记录中的交易数量更正为正数，更新item表。

```python
#表中筛选出ITEM_QUANTITY小于0的异常值
item[item['ITEM_QUANTITY']<0]
```

```python
#处理商品数量值为负的异常值，改为正数,更新item表
item['ITEM_QUANTITY']=abs(item['ITEM_QUANTITY'])
#item
#item[item['ITEM_QUANTITY']<0]
```

### 2.实际交易金额问题

<font color=black size=4>&emsp;&emsp;各商店实际交易金额并存在与以上与交易数量相同的问题，即实际交易金额存在负数的情况为2020条明细，观察处理数据发现交易数量与实际交易金额均为负值可能导致异常实际折扣数值异常，因此本次在python中对存在负值问题的数据进行了绝对值处理，将其同样更正为正值。 
    
<font color=black size=4>&emsp;&emsp;如下表所示，通过数据显示在交易记录明细当中有636条交易记录的实际交易金额为0元，且折扣类型都为Q。且对交易日期重新排序后显示，记录在册的各超市商店在2018年6月1日—2018年7月4日期间对部分商品进行全折销售,7月5日各商店免费活动结束。

```python
#将实际交易金额小于0的异常记录更正，负值更正为正值
item['ITEM_SPEND']=abs(item['ITEM_SPEND'])
#item[item['ITEM_SPEND']<0]
```

```python
#筛选出实际交易金额等于0的交易记录，发现折扣类型为q，有633条商品进行免费赠送
item_freespend=item[item['ITEM_SPEND']==0]
item_freespend['SDATE']=sorted(item_freespend['SDATE'])
item_freespend
```

### 3.折扣金额问题

<font color=black size=4>&emsp;&emsp;由于交易数量以及实际交易金额出现问题，本次对折扣金额也同样做出对比计算来验证再计算折扣金额和实际各商店记录折扣金额是否存在出入。本次通过简单计算编写算式表达式重新计算折扣金额，表达式为实际交易金额=交易数量*正常交易金额，且按照大多数商店计数规则保留2位小数。
对前步骤更改过后，结果对比发现在1732761项交易记录明细当中有6524条记录存在不对等的情况。经过初步分析导致DISCOUNT_AMOUNT和RDISCOUNT_AMOUNT不对等的原因有三类对于异常值的类型不同，本报告处理数据的方式也不尽相同，具体如下所示：

```python
#重新计算各商品折扣价格，添加列REDISCOUNT_AMOUNT
item['REDISCOUNT_AMOUNT']=round(item['ITEM_SPEND']-round(item['ITEM_QUANTITY']*item['NORMAL_PRICE'],2),2)
#item
#diff1 = set(item['REDISCOUNT_AMOUNT']).difference(item['DISCOUNT_AMOUNT'])
#item['diff']=item['REDISCOUNT_AMOUNT']-item['DISCOUNT_AMOUNT'

# DISCOUNT_AMOUNT和RDISCOUNT_AMOUNT不对等的记录
differentdiscount= item.loc[(item['DISCOUNT_AMOUNT']!=item['REDISCOUNT_AMOUNT'])]
differentdiscount
```

<font color=black size=4>&emsp;&emsp;数据交易明细中DISCOUNT_AMOUNT存在两种记录方式，例如折扣价格-40和40元，因此在数据进入下一步分析应统一记录方式。下列显示了具体处理此类数据的过程，将折扣金额统一以负值的形式记录。


```python
differentdiscount.iloc[6523]
```

```python
#将DISCOUNT_AMOUNT改为负值的折扣记录方式，并重新筛选出DISCOUNT_AMOUNT和REDISCOUNT_AMOUNT不对等的交易记录
item["DISCOUNT_AMOUNT"]=abs(item["DISCOUNT_AMOUNT"])
item["DISCOUNT_AMOUNT"]= -item["DISCOUNT_AMOUNT"]
item["REDISCOUNT_AMOUNT"]=abs(item["REDISCOUNT_AMOUNT"])
item["REDISCOUNT_AMOUNT"]=-item["REDISCOUNT_AMOUNT"]
differentdiscount1= item.loc[(item['DISCOUNT_AMOUNT']!=item['REDISCOUNT_AMOUNT'])]
differentdiscount1
```

   （1）原始交易明细数据记录的为2018年6月1日-7月5日的顾客购买商品的交易情况。35天内商店交易数量以“A”字形上下波动。交易开始记录的一个星期内客流量下降，交易数量降低，直到下周开始有所回升，直到直到2018年6月17日左右到达顶峰，而19日出现了日交易数量最低点，低估过后日交易次数再次回升，但总体数量明显低于出现日高峰前的交易数量。
   
   （2）交易明细记录显示超市连锁公共有24家分店，编码为A 的商店有15家，编码为D的商店有9家。35天内总体销售量最高的商店分别D002和A001销售数量最低的门店为D003，从上述分析结果可以看到各门店的销售水平参差不齐，绝大多数的门店销售量仅占总量的5%-10%之间。
    
   （3）使用率最高的POS机为1号、2号和3号POS机，所有商店有281042个不同的购物篮，且若将商品销出数量定为A-E等级，发现E等级商品占比为90.66%，较受欢迎的商品仅占10%。各个门店销售前三的商品也可以通过统计筛选的手段统计得出。
    
   （4）通过计算交易数量、实际交易金额及折扣金额可以得出，实际交易金额与折扣金额间存在一定的线性负相关关系，当实际交易金额增加时，折扣金额呈现下降分布。

### 2.建议

<font color=black size=4>&emsp;&emsp;本文针对以上分析结论及交易明细体现出商店的销售问题，提出以下几点结论：
    （1）商品日交易水平参差不齐，商店应从以上趋势变动及数据分析做好对商品需求的管控工作，在日交易高峰期低谷期时，超市应及时处理滞销商品，高峰期时，保证热销商品的供应，提前进行合理规划。 
    （2）商店大部分门店销售量较低，超市应采取更加有效的管理策略及运营手段来增加销售量高的商品占比以此提高盈利利润，例如各商店联动活动、商品促销等经营手段。
    （3）商店进购的商品大部分为销售冷门商品，超市应尽快核对冷门商品，减少进货或替代以达到销售利润提高的目的。
